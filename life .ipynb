```python
most_10_countries = sus.groupby('Country')['Suicides number'].sum().sort_values(ascending=False).reset_index()
most_10_countries = most_10_countries.head(10)
most_10_countries
```

```python
plt.figure(figsize=(10, 6))
barplot = sns.barplot(x='Country', y='Suicides number', data=most_10_countries)
plt.xlabel('Country')
plt.ylabel('Total Suicides')
plt.xticks(rotation=60)
plt.title('Top 10 Countries by Total Suicides(2000 - 2011)')
plt.show()
```

```python
avg_10_countries = sus.groupby('Country')['GDP'].mean().sort_values(ascending=False).reset_index()
avg_10_countries=avg_10_countries.head(10)
avg_10_countries
```

```python
plt.figure(figsize=(8, 8))
plt.pie(avg_10_countries['GDP'],labels=avg_10_countries['Country'],autopct='%1.1f%%')
plt.title('avrage GDP per country (top 10)')
plt.show()
```

```python
def convert_nos(z):
    for column in z.columns[~z.columns.isin(['Year', 'Suicides number', 'Life expectancy',
       'Adult Mortality', 'Infant deaths', 'Alcohol', 'Under-five deaths',
       'HIV/AIDS', 'GDP', 'Population', 'Income composition of resources',
       'Schooling'])]:
        z[column]=z[column].factorize()[0]
    return z  
```

```python
sus_encoded=convert_nos(sus)
```

```python
sus_encoded
```

```python
sus_encoded.corr()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(sus_encoded.corr(),annot=True,cmap='rainbow')
```
